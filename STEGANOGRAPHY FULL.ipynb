```python
#STEGANOGRAPHY FULL
#ENCRYPTION
print("ENTER ALPHABETS AND NUMBERS or !#$%&'()*+""")
a=input("ENTER THE PLAIN TEXT: ")
b=int(input("ENTER A VALUE "))
d=len(a)
st=[]
for i in range(d):
    s=a[i]
    if(s>="!" and s<="+"):
        m=chr(((ord(s)-33+b)%10)+33)
        st.append(m)
    elif(s>="0" and s<="9"):
        m=chr(((ord(s)-48+b)%10)+48)
        st.append(m)
    elif(s==s.upper()):
        m=chr(((ord(s)-65+b)%26)+65)
        st.append(m)
    elif(s==s.lower()):
        m=chr(((ord(s)-97+b)%26)+97)
        st.append(m)  
print("The cipher text is ")
for k in st:
    print(k,end="")
```
